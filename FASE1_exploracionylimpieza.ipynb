2. Limpieza de datos:

```python
#Eliminamos duplicados
```

```python
#Cambiamos los nombre de las columnas a minúsculas.

nuevas_columnas = {columna: columna.lower() for columna in df_unido.columns}
df_unido.rename (columns = nuevas_columnas, inplace = True)
df_unido.head(1)
```

```python
#Convertimos de FLOAT a INT

def convertir_float_int(columna):
    try:
        numero = int(columna)
        return numero
    except(ValueError, TypeError):
        return np.nan
```

```python
#Se entiende que no existen medios puntos.
df_unido["points accumulated"] = df_unido["points accumulated"].apply(convertir_float_int)
```

```python
#Se convierte a INT porque los años, ni los meses pueden ser FLOAT. La función no cambiaba los datos y he elegido esta forma. 
df_unido["cancellation year"] = df_unido["cancellation year"].astype('Int64')
```

```python
df_unido["cancellation month"] = df_unido["cancellation month"].astype('Int64')
```

```python
def convertir_int_float(columna):
    try:
        numero = float(columna)
        return numero
    except:
        return np.nan
```

```python
#def convertir_nulos(columna):
    #if columna < 0:
        #return np.nan
    #else:
        #return columna
```

```python
#Convertimos esta columna a float porque es una moneda
df_unido["dollar cost points redeemed"] = df_unido["dollar cost points redeemed"].apply(convertir_int_float)
```

```python
#Cambiamos los negativos por nulos
valores_menores_que_cero = df_unido[df_unido['salary'] < 0]['salary']
len(valores_menores_que_cero)
```

```python
df_unido.loc[df_unido['salary'] < 0, 'salary'] = np.nan
```

```python
valores_menores_que_cero = df_unido[df_unido['salary'] < 0]['salary']
len(valores_menores_que_cero)
```

```python
#Ahora necesitamos crear una columna numerica para education que nos permita usarla en el iterative
df_unido['education'].unique()
```

```python
def convertir_numerica(valor):
    if valor == 'High School or Below':
        return 1
    elif valor == 'Bachelor':
        return 2
    elif valor == 'College':
        return 3
    elif valor == 'Master':
        return 4
    elif valor == 'Doctor':
        return 5
```

```python
df_unido.insert(16, 'education_num', df_unido["education"].apply(convertir_numerica))
```

```python
df_unido.dtypes
```

```python
#Nulos:
df_unido.isnull().sum()
```

```python
nulos_porcentaje_salario = (df_unido['salary'].isnull().sum()/df_unido.shape[0])*100
nulos_porcentaje_salario
```

```python
df_copia = df_unido.copy()
```

```python
imputer_iterative = IterativeImputer(max_iter = 20, random_state = 42)

# ajustamos y tranformamos los datos
imputer_iterative_imputado = imputer_iterative.fit_transform(df_copia[["salary", "education_num", 'clv']])

# comprobamos que es lo que nos devuelve, que en este caso es un array también
imputer_iterative_imputado
```

```python
df_copia[["salary_iterative", "education_iterative", 'clv_iterative']] = imputer_iterative_imputado
```

```python
print(f"Después del 'Iterative' tenemos: \n{df_copia[['salary_iterative', 'education_iterative', 'clv_iterative']].isnull().sum()} nulos")
```

```python
# instanciamos la clase del KNNImputer
imputer_knn = KNNImputer(n_neighbors = 5)

# ajustamos y transformamos los datos
imputer_knn_imputado = imputer_knn.fit_transform(df_copia[["salary", "education_num", 'clv']])

# comprobamos que es lo que nos devuelve, que sigue siendo un array
imputer_knn_imputado
```

```python
df_copia[["salary_knn", "education_knn", "clv_knn"]] = imputer_knn_imputado


# comprobamos los nulos
print(f"Después del 'KNN' tenemos: \n{df_copia[['salary_knn', 'education_knn', 'clv_knn']].isnull().sum()} nulos")
```

```python
df_copia.drop(["salary","salary_iterative", "education_num", "education_iterative", "education_knn", "clv", "clv_iterative", "clv_knn"], axis = 1, inplace = True)
```
